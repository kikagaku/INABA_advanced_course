### Q4

---

以下のデータセットについて、多くの値は `United-States` となっており、その他の国に関してはあまり存在していません。こちらの列について、以下のように変換を行ってください。

- income.csv

**処理内容**

- United-States を 0 と変換
- 欠損値と ? は 1 に変換
- それ以外の値を 2 に変換

- ヒント
 - ? を nan に変換する際には `df = df.replace(' ?', np.nan)` で置き換えることができます



### Q5

- income.csv

**処理内容**

### Q7

---

以下のデータの `Timestamp` 列には日付データと時間のデータが同じ列に含まれています。

- network_analytics.csv

以下のように変換してください。

**変換前**

| Timestamp | OutboundUtilzation (%) |
| --- | --- |
| 10/10/2017 7:01 | 96.2442 |
| 10/10/2017 7:21 | 99.1131 |

**変換後**

| Timestamp | Time | OutboundUtilzation (%) |
| --- | --- | --- |
| 10/10/2017 | 7:01 | 96.2442 |
| 10/10/2017 | 7:21 | 99.1131 |

- ヒント：データの分割は `split`メソッドを使用します。`df = df['変数'].str.split(' ', expand=True)`
